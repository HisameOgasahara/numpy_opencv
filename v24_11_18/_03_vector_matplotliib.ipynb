```python
import numpy as np
import matplotlib.pyplot as plt

# 벡터 [4 3]^T
a = np.array([4, 3])


fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1)

# 벡터를 화살표로 나타냅니다.
# 시작점은 원점, 끝점은 벡터 [4 3]^T 입니다. 
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], 'a', size=15)


ax.set_xticks(range(0, 7))
ax.set_yticks(range(0, 7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


a = np.array([4, 1])
b = np.array([-2, 3])
sum = a + b


fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1)


# 벡터 a를 원점에서 시작하도록 그립니다.
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], "a", size=15)

# 벡터 b를 벡터 a의 끝점에서 시작하도록 그립니다.
ax.quiver(a[0], a[1], b[0], b[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0]+b[0], a[1]+b[1], "b", size=15)

# 벡터의 합을 그리면 벡터 a의 시작점과 벡터 b의 끝점을 잇는 벡터가 그려집니다.
ax.quiver(0, 0, sum[0], sum[1], angles='xy', scale_units='xy', scale=1, color='blue')
ax.text(sum[0]*0.5+0.3, sum[1]*0.5, "a+b", size=15, color='blue')


ax.set_xticks(range(0, 7))
ax.set_yticks(range(0, 7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')


ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# 벡터 A와 B를 생성합니다. 
A = np.array([1, -2])
B = np.array([4, 3])


# 1개의 행과 2개의 열을 가지는 서브플롯을 생성합니다. 
f, ax= plt.subplots(1, 2)

# 다음처럼 인덱스를 사용하여 각 서브플롯을 사용합니다.
# 각 서브플롯 상단에 텍스트 추가합니다.
ax[0].title.set_text('A')
ax[1].title.set_text('B')


# 각 서브플롯에 시작점이 (0,0)인 벡터를 그립니다.
ax[0].quiver(0, 0, A[0], A[1], color = 'red', angles='xy', scale_units='xy', scale=1)
ax[1].quiver(0, 0, B[0], B[1], color = 'blue', angles='xy', scale_units='xy', scale=1)

# 서브플롯별로 그리드를 생성합니다.
start_x = -1
end_x = 5
start_y = -3
end_y = 4
for i in range(2):
    ax[i].axis([start_x, end_x, start_y, end_y])
    ax[i].set_xticks(range(start_x, end_x))
    ax[i].set_yticks(range(start_y, end_y))
    ax[i].grid(True)
    ax[i].set_axisbelow(True)
    ax[i].set_aspect('equal', adjustable='box')
    ax[i].spines['left'].set_position('zero')
    ax[i].spines['bottom'].set_position('zero')
    ax[i].spines['right'].set_color('none')
    ax[i].spines['top'].set_color('none')


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1개의 행과 2개의 열을 가지는 서브플롯을 생성합니다. 
fig, ax= plt.subplots(1, 2, figsize=(10,16))
a = np.array([4, 1])
b = np.array([-2, 3])

axis_info=[{'sum':a+b, 'sign':1, 'title':'A+B', 'color':'red'},
           {'sum':a-b, 'sign':-1,'title':'A-B', 'color':'blue'}]

# 서브플롯별로 그리드를 생성합니다.
start_x = 0
end_x = 7
start_y = -3
end_y = 5
for i in range(2):
    ax[i].title.set_text(axis_info[i]['title'])
    ax[i].quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
    ax[i].text(a[0], a[1], "a", size=10)
    tmp = axis_info[i]['sign']
    str_tmp = '+' if tmp > 0 else '-'
    ax[i].quiver(a[0], a[1], tmp*b[0], tmp*b[1], angles='xy', scale_units='xy', scale=1)
    ax[i].text(a[0]+tmp*b[0]+0.1, a[1]+tmp*b[1]+0.1, f"{str_tmp}b", size=10)
    ax[i].quiver(0, 0, axis_info[i]['sum'][0], axis_info[i]['sum'][1], 
                 color=axis_info[i]['color'], angles='xy', scale_units='xy', scale=1)
    ax[i].text(axis_info[i]['sum'][0]*0.5+0.3, axis_info[i]['sum'][1]*0.5, 
               f"a{str_tmp}b", size=15, color=axis_info[i]['color'])
    
    ax[i].set_xticks(range(start_x, end_x+1)) # x축 (시작,끝-1) 설정
    ax[i].set_yticks(range(start_y, end_y+1)) # y축 (시작,끝-1) 설정
    ax[i].grid(True)                        # 그리드 표시
    ax[i].set_axisbelow(True)               # 그리드 객체 아래로
    ax[i].set_aspect('equal', adjustable='box') # 가로세로비율 1:1
    ax[i].spines['left'].set_position('zero')   # 좌표축 x = 0
    ax[i].spines['bottom'].set_position('zero') # 좌표축 y = 0
    ax[i].spines['right'].set_color('none')     # 오른쪽 외곽선 없음
    ax[i].spines['top'].set_color('none')       # 위쪽 외곽선 없음
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


a = np.array([4, 1])
c = 3
ca = np.multiply(c, a)


fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1)


# 벡터 a를 그립니다.
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
plt.text(a[0], a[1], 'a', size=15)

# 벡터 a에 3을 곱한 결과를 그립니다. 시작점을 (0,1)로 이동하여 그렸습니다. 
plt.quiver(0, 1, ca[0], ca[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.text(ca[0], ca[1]+1, '3a', size=15, color='blue')


ax.set_xticks(range(-2, 15))
ax.set_yticks(range(-3, 9))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')


ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data generation
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# 3D surface plot creation
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('2-variable function z = sin(x)cos(y)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data generation
x = np.arange(-5, 5, 1)
y = np.arange(-5, 5, 1)
X, Y = np.meshgrid(x, y)

# Map application
U = X + Y
V = X - Y

# Vector field plot creation
plt.figure()
plt.quiver(X, Y, U, V)
plt.xlabel('x')
plt.ylabel('y')
plt.title('2-variable map (x, y) -> (x + y, x - y)')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```
