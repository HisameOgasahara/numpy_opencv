```python
# 배열 결합 함수

# 두 행렬을 세로(열방향)로 붙임 np.vstack(A, B) <=> np.concanate((A,B), axis=0)
# 두 행렬을 가로(행방향)로 붙임 np.hstack(A,B) <=> np.concanate((A,B),axis=1)
# np.column_stack((A,B,C)) : 1차원 배열을 열벡터로 하는 2차원 배열 생성. 세로 막대기 여러개를 붙임.

A = np.arange(1,5).reshape(2,2)
print(A)

print()

B = np.eye(2,dtype=np.int32) #단위행렬
print(B)

#가로 방향(행방향)으로 합치기
C = np.hstack((A,B))
print(C)

print()

D = np.concatenate((A,B), axis=1)
print(D)

print()

E = np.vstack((A,B))
print(E)

print()

F = np.concatenate((A,B), axis=0)
print(F)
```

```python
# import numpy as np

A = np.arange(0, 15, 2)
for i in range(len(A)):
    print(f"A[{i}]:{A[i]}", end="")

print()
print('A[0:3]:', A[0:3]) #인덱스 0부터 3앞까지
print('A[0:4]:', A[0:4]) # 인덱스 0부터 4앞까지
print('A[:4]:', A[:4]) # 인덱스 4앞까지 
print("A[6:8]:", A[6:8]) #인덱스 6부터 8(인덱스 8은 없음. 즉 끝까지)
print("A[6:]:", A[6:]) # 인덱스 6부터 끝가지
print("A[:]:", A[:]) #전체 원소 포함
print("A[::2]:", A[::2]) #첫번째 원소부터 2씩 건너뛰기
print("A[:-2]:", A[:-2]) #첫번째 마이너스 인덱스(-8)부터 -2앞까지
print("A[-2:]:", A[-2:]) #끝에서 2번재(-2)부터 끝까지
```

```python
# 대입 / 얕은 복사(reshape) / 깊은 복사(copy) : 사본이 필요하면 copy를 하자!


# 대입 : 넘파이 배열을 가리키는 변수를 다른 변수에 대입, 2개의 변수가 하나의 배열을 가리킴
# 얕은 복사 : shape변경 함수(reshape)등을 사용하여 원본 데이터 공유. 원본 변경시 데이터 함께 변경
# 깊은 복사 : copy()를 사용하여 완전한 복사본 생성, 원본과 복사본은 관련 없음 


###대입

A = np.arange(1,13)
print(A)
B = A #대입
print(B)
print(A==B) #값이 같니? 행렬 각 원소마다 비교
print(A is B) # 같은 메모리 주소 참조? YES

A[0] = 100
B[-2] = -11
print(A) 
print(B) #서로 영향을 준다
```

```python
### 얕은 복사. 복사한다 싶으면 이거 말고 밑에 copy()를 쓰자

A = np.arange(1,13)
print(A)
B = A.view() #얕은 복사
print(B)
print(B.base is A) #같은 base를 사용
print(B is A) #메모리는 다름.

#원본을 바꾸면 얕은복사 B도 바뀜
A[0] = 100
print(A)
print(B)

print()

#사본을 바꿔도 원본 A도 바뀐다
B[1] = -3
print(A)
print(B)
```

```python
###깊은 복사 : copy()사용, 사본이 생성되면 원본과 상관없음. 


A = np.arange(1,13)
print(A)
B = A.copy() #얕은 복사
print(B)

print()

A[0] = 100
B[1] = -3
print(A)
print(B)
```
